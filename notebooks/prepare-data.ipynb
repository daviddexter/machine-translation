```python
from igbo_text import IgboText
import numpy as np
import pandas as pd
from keras.preprocessing.text import text_to_word_sequence
from collections import Counter
```

```python
igbo_text = IgboText()
dataset = pd.read_csv('../data/data.csv')
```

```python
tokenized_igbo = []
```

```python
texts = dataset['ig'].tolist()
```

```python
len(texts)
```

```python
for text in texts:
    text = " ".join(text_to_word_sequence(text, filters='!#$%&()*+,./:;<=>?@[\\]^_`{|}~\t\n', lower=False, split=' '))
    if text.strip() == "":
        continue
    tokenized_igbo += igbo_text.tokenize(text, convert_to_lower=True)
```

```python
igbo_counter = Counter(tokenized_igbo)
```

```python
len(igbo_counter)
```

```python
igbo_counter.most_common(40)
```

```python
texts = dataset['en'].tolist()
```

```python
len(texts)
```

```python
tokenized_eng = []
```

```python
for text in texts:
    tokenized_eng += text_to_word_sequence(text)
```

```python
english_counter = Counter(tokenized_eng)
```

```python
len(english_counter)
```

```python
english_counter.most_common(40)
```
